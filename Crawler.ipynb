```
import requests
from bs4 import BeautifulSoup
```

```
List=['thoi-su','kinh-doanh','the-gioi','giai-tri','the-thao','phap-luat','giao-duc','doi-song','du-lich','khoa-hoc','so-hoa','oto-xe-may','y-kien','hai']
```

```
def check_date(k):
  a=k.split(',')
  b=a[1].split('/')
  if int(b[-1]) == 2020:
    if int(b[-2]) ==  9 or (int(b[-2]) == 8 and int(b[-3])>20):
      return 1
```

```
def crawl_date(url, date):
  r1 = requests.get(url)
  coverpage = r1.content
  soup1 = BeautifulSoup(coverpage, 'html5lib')
  coverpage_news = soup1.find_all("span", class_="date")
  if coverpage_news == []:
    return None
  for i in coverpage_news:
    return str(i.get_text())
```

```
def crawl_data(url, data):
  r1 = requests.get(url)
  coverpage = r1.content
  soup1 = BeautifulSoup(coverpage, 'html5lib')
  coverpage_news = soup1.find_all("h1", class_="title-detail")
  if coverpage_news == []:
    return None
  for i in coverpage_news:
    return i.get_text()
```

```
def crawl_url(url, html):
  r1 = requests.get(url)
  coverpage = r1.content
  soup1 = BeautifulSoup(coverpage, 'html5lib')
  coverpage_news = soup1.find_all("a")
  if coverpage_news == []:
    return None
  for i in coverpage_news:
    a = str(i.get("href"))
    if a.endswith('html') and a not in html:
     html.append(i.get("href"))
```

```
url = "https://vnexpress.net/"
html= []
date=[]
link=[]
data=[]
category=[]
temp=0
t=1
for i in List:
  print(i)
  while temp<2:
    if i=='kinh-doanh' or i=='the-thao' or i=='suc-khoe' or i=='doi-song' or i=='du-lich' or i=='so-hoa':
      if t==1:
        URL=url+i
      else:
        URL=url+str(i)+"/p"+str(t)
    else:
      URL=url+str(i)+"-p"+str(t)
    print(URL)
    crawl_url(URL, html)
    r1 = requests.get(url)
    coverpage = r1.content
    soup1 = BeautifulSoup(coverpage, 'html5lib')
    for y in html:
      s=crawl_date(y,date)
      if s!=None:
        if check_date(s) ==1:
          date.append(s)
          x=crawl_data(y,data)
          data.append(x)
          link.append(y)
          category.append(i)
          print(len(date),len(link),len(data),len(category))
        else:
          temp=temp+1
          print(temp,': bien dem')
    t=int(t)
    t+=1
    html.clear()
  t=1
  temp=0
```
